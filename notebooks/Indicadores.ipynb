```python
df_activitat = pd.read_csv("df_activitat.csv")
```

```python
df_activitat
```

```python
df_all = df_all.merge(df_activitat, on='nom_activitat', how='left')
df_all
```

```python
def select_cols(df):
    columns = ['id_barri',
               'id_activitat_y',
              ]

    df = df[columns]

    return df
```

```python
df_all_clean = select_cols(df_all)
df_all_clean
```

```python
df_agrupado = df_all_clean.groupby('id_barri')['id_activitat_y'].value_counts().reset_index(name='conteo')
df_agrupado
```

```python
df_agrupado = df_agrupado.rename(columns={
                                    'id_barri': 'id_barri',
                                    'id_activitat_y': 'id_activitat',
                                    'conteo': 'total_activitat',
})
```

```python
#pd.options.display.max_rows = None
df_agrupado
```

```python
df_agrupado['percent_total'] = df_agrupado['total_activitat'] / df_agrupado.groupby('id_activitat')['total_activitat'].transform('sum') * 100
df_agrupado['percent_total'] = df_agrupado['percent_total'].round(2)
df_agrupado
```

```python
df_agrupado['percent_rel'] = ((1-((df_agrupado['percent_total'] / (100/73))))*100)*(-1)
df_agrupado
```

```python
df_filtrado = df_agrupado[df_agrupado['id_activitat'] == 9]
df_filtrado
```

```python
wiki = pd.read_html('https://ca.wikipedia.org/wiki/Districtes_i_barris_de_Barcelona')[0]
wiki
```

```python
wiki_clean = wiki['Barris[4]'][['Núm', 'Població']]
```

```python
wiki_clean.columns = ['id_barri', 'població']
```

```python
wiki_clean['id_barri'] = pd.to_numeric(wiki_clean['id_barri'], errors='coerce')
wiki_clean
```

```python
df_poblacio = df_agrupado.merge(wiki_clean, on='id_barri', how='left')
df_poblacio
```

```python
df_poblacio['població'] = df_poblacio['població'].str.replace('.', '').astype(int)
```

```python
df_poblacio['activ_hab_abs'] = df_poblacio['població']/df_poblacio['total_activitat']
df_poblacio
```

```python
df_poblacio['activ_hab_rel'] = df_poblacio['total_activitat'] / (df_poblacio.groupby('id_activitat')['total_activitat'].transform('sum')/73)
df_poblacio
```

```python
df_poblacio = df_poblacio.drop(['percent_total', 'percent_rel'], axis=1)
df_poblacio
```

```python
En el DataFrame de "indicadores" hay:
- població: poblacion por barrio
- activ_hab_abs: 
    Es el número de actividades (valor absoluto) que hay por población en un barrio en concreto.
    Por ejemplo en la fila 0, si la actividad con id = 9 fueran "farmacias", significa que hay 1 farmacia por cada 80 personas.
- activ_hab_rel: es el número de actividades que hay por población en un barrio en concreto.
    Es
    Por ejemplo en la fila 0, si la actividad con id = 9 fueran "farmacias", significa que en ese barrio hay 2.5 veces el doble de farmacias que la media de todos los barrios. 
    
    
-
-
```
