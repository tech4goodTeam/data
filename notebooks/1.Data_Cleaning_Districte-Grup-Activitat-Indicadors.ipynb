## rename cols

```python
def rename_cols(df):
    """
    Aquesta funció canvia de nom les columnes del df. 
    Converteix les majúscules en minúscules i posa un guió baix entre paraules en cas de no haver-n'hi.
    """
    cols = []
    for i in range(len(df.columns)):
        cols.append(df.columns[i].lower().replace(' ', '_'))
    df.columns = cols
    return df
```

```python
df = rename_cols(df)
```

```python
def clean_nom_activitat(df):
    """
    Aquesta funció realitza agrupacions i reemplaçaments en les columnes 'nom_activitat' i 'nom_grup_activitat'.
    """
    df['nom_activitat'] = df['nom_activitat'].replace(['Altres', 'Sense informació', 'Sense informació - Havia tingut activitat (rètol)', 'altres', 'Altres ( per exemple VENDING)'], 'Altres')
    df['nom_activitat'] = df['nom_activitat'].replace(['Basars', 'Souvenirs', 'Souvenirs i basars'], 'Souvenirs i basars')
    df['nom_activitat'] = df['nom_activitat'].replace(['Altres equipaments esportius', 'Esports'], 'Equipaments esportius / Esports')
    df['nom_activitat'] = df['nom_activitat'].replace(['Òptiques i fotografia', 'Fotografia'], 'Fotografía')
    df['nom_activitat'] = df['nom_activitat'].replace(['serveis de menjar i begudes', 'Serveis de menjar take away MENJAR RÀPID'], 'Serveis menjar i begudes per emportar')


    df['nom_activitat'] = df['nom_activitat'].replace('Plats preparats (no degustació)', 'Plats preparats')
    df['nom_activitat'] = df['nom_activitat'].replace('Farmàcies PARAFARMÀCIA', 'Farmàcies / Parafarmácies')
    df['nom_activitat'] = df['nom_activitat'].replace('Bars   / CIBERCAFÈ', 'Bars / Cibercafé')
    df['nom_activitat'] = df['nom_activitat'].replace("serveis d'allotjament", "Serveis d'allotjament")
    df['nom_activitat'] = df['nom_activitat'].replace('Gimnàs /fitnes', 'Gimnás / Fitness')
    df['nom_activitat'] = df['nom_activitat'].replace('Herbolaris, dietètica i NUTRICIÓ', 'Herbolaris, dietètica i nutrició')
    
    
    df['nom_grup_activitat'] = df['nom_grup_activitat'].replace("Restaurants, bars i hotels (Inclòs hostals, pensions i fondes)", "Restaurants, bars, hotels i hostals")
    df['nom_grup_activitat'] = df['nom_grup_activitat'].replace(['Grup no definit', 'Altres'], 'Altres')
    return df
```

# Select  & Rename cols

```python
def select_rename_cols(df):
    """
    Aquesta funció selecciona columnes específiques d'un DataFrame 
    i realitza una sèrie de renomenaments en aquestes columnes.
    """
    columns = [ 'codi_grup_activitat', 
                'nom_grup_activitat',  
               
                'codi_activitat_2019',
                'nom_activitat',
               
                'codi_barri',
                'nom_barri',
               
                'codi_districte',
                'nom_districte']

    df = df[columns]

    df = df.rename(columns={
    'codi_grup_activitat': 'id_grup',
    'nom_grup_activitat': 'nom_grup',
        
    'codi_activitat_2019': 'id_activitat',
    'nom_activitat': 'nom_activitat',
        
    'codi_barri': 'id_barri',
    'nom_barri': 'nom_barri',
        
    'codi_districte': 'id_districte',
    'nom_districte': 'nom_districte'
    })

    return df
```

```python
df = select_rename_cols(df)
```

## Indicadores

```python
df_all = pd.read_csv("../data/all_clean.csv")
df_all
```

```python
df_all = df_all.merge(df_activitat_unique, on='nom_activitat', how='left')
```

```python
def select_cols(df):
    columns = ['id_barri','id_activitat_y',]
    df = df[columns]
    df = df.rename(columns={'id_barri': 'id_barri', 'id_activitat_y': 'id_activitat'})

    return df
```

```python
df_indicadors = select_cols(df_all)
df_indicadors
```

```python
df_indicadors = df_indicadors.groupby('id_barri')['id_activitat'].value_counts().reset_index(name='total_activitat')
df_indicadors
```

### Població - Wikipedia

```python
wiki = pd.read_html('https://ca.wikipedia.org/wiki/Districtes_i_barris_de_Barcelona')[0]
wiki
```

```python
wiki_clean = wiki['Barris[4]'][['Núm', 'Població']]
```

```python
wiki_clean.columns = ['id_barri', 'població']
```

```python
wiki_clean['id_barri'] = pd.to_numeric(wiki_clean['id_barri'], errors='coerce')
wiki_clean
```

### Població

```python
df_indicadors_poblacio = df_indicadors.merge(wiki_clean, on='id_barri', how='left')
df_indicadors_poblacio
```

```python
df_indicadors_poblacio['població'] = df_indicadors_poblacio['població'].str.replace('.', '').astype(int)
```

```python
df_indicadors_poblacio['activ_hab_abs'] = df_indicadors_poblacio['població']/df_indicadors_poblacio['total_activitat']
df_indicadors_poblacio
```

```python
df_indicadors_poblacio['activ_hab_rel'] = df_indicadors_poblacio['total_activitat'] / (df_indicadors_poblacio.groupby('id_activitat')['total_activitat'].transform('sum')/73)
df_indicadors_poblacio
```
